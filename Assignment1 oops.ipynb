```python
class BankAccount:
    def __init__(self , account_number , account_holder_name , balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance
        
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f" deposited {amount} successfully . New balance {self.balance}")
        else:
            print("ivalid deposit amount")
                  
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"you have successfully withdrawn {amount} . New balance {self.balance}")
        else:
            print("insufficient balance or invalid amount")

                        
account1 = BankAccount("123456" , "VENKATRAM" , 31000)
print("initial balance :" , account1.balance)
account1.deposit(5000)
account1.withdraw(4000)
account1.withdraw(30000)
account1.withdraw(5000)
                        
```

```python
class Employee:
    def __init__(self , employee_id , name , salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary
        
    def calculate_yearly_bonus(self , bonus_percentage):
        bonus = (bonus_percentage / 100) * self.salary
        return bonus
    
    def display_details(self):
        print("Employee ID:" , self.employee_id)
        print("Name:" , self.name)
        print("Salary:" , self.salary)
        
employee1 = Employee(101,81,50000) 
employee1.display_details()  
bonus = employee1.calculate_yearly_bonus(10)
print("Yearly Bonus:" , bonus)
```

```python
class Vehicle:
    def __init__(self, vehicle_id, brand, model, year, is_available=True):
        self.vehicle_id = vehicle_id
        self.brand = brand
        self.model = model
        self.year = year
        self.is_available = is_available

    def __str__(self):
        return f"{self.brand} {self.model} ({self.year}) - Available: {self.is_available}"


class RentalSystem:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def rent_vehicle(self, vehicle_id):
        for vehicle in self.vehicles:
            if vehicle.vehicle_id == vehicle_id:
                if vehicle.is_available:
                    vehicle.is_available = False
                    print(f"Vehicle {vehicle_id} rented successfully.")
                else:
                    print(f"Vehicle {vehicle_id} is not available for rent.")
                return
        print(f"Vehicle {vehicle_id} not found.")

    def return_vehicle(self, vehicle_id):
        for vehicle in self.vehicles:
            if vehicle.vehicle_id == vehicle_id:
                vehicle.is_available = True
                print(f"Vehicle {vehicle_id} returned successfully.")
                return
        print(f"Vehicle {vehicle_id} not found.")

    def display_available_vehicles(self):
        available_vehicles = [vehicle for vehicle in self.vehicles if vehicle.is_available]
        if available_vehicles:
            print("Available Vehicles:")
            for vehicle in available_vehicles:
                print(vehicle)
        else:
            print("No vehicles available for rent.")



car1 = Vehicle("001", "Toyota", "Camry", 2020)
car2 = Vehicle("002", "Honda", "Accord", 2019)
car3 = Vehicle("003", "Ford", "Fusion", 2018)

rental_system = RentalSystem()
rental_system.add_vehicle(car1)
rental_system.add_vehicle(car2)
rental_system.add_vehicle(car3)

rental_system.rent_vehicle("001")
rental_system.display_available_vehicles()

rental_system.return_vehicle("001")
rental_system.display_available_vehicles()
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            return True
        else:
            print("Sorry, this book is already borrowed.")
            return False

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print("Thank you for returning the book.")
        else:
            print("This book was not borrowed.")


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_available_books(self):
        available_books = [book for book in self.books if not book.is_borrowed]
        if available_books:
            print("Available books in the library:")
            for book in available_books:
                print(f"{book.title} by {book.author}")
        else:
            print("There are no available books in the library.")

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title:
                if book.borrow():
                    print(f"You have borrowed '{book.title}' by {book.author}.")
                return
        print("Sorry, the requested book is not available in the library.")



if __name__ == "__main__":
    
    book1 = Book("Harry Potter and the Sorcerer's Stone", "J.K. Rowling")
    book2 = Book("The Great Gatsby", "F. Scott Fitzgerald")
    book3 = Book("To Kill a Mockingbird", "Harper Lee")


    library = Library()

    
    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)

    
    library.display_available_books()

    
    library.borrow_book("The Great Gatsby")
    library.borrow_book("The Catcher in the Rye")

    
    library.display_available_books()

    
    book2.return_book()


    library.display_available_books()
```

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        self.quantity = new_quantity

    def __str__(self):
        return f"{self.name} - Price: ${self.price}, Quantity: {self.quantity}"


class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def update_product_quantity(self, product_name, new_quantity):
        for product in self.products:
            if product.name == product_name:
                product.update_quantity(new_quantity)
                print(f"Quantity of '{product_name}' updated to {new_quantity}.")
                return
        print(f"Product '{product_name}' not found in the inventory.")

    def display_available_products(self):
        if self.products:
            print("Available products in the inventory:")
            for product in self.products:
                print(product)
        else:
            print("Inventory is empty.")



if __name__ == "__main__":
    
    product1 = Product("Laptop", 1000, 10)
    product2 = Product("Mouse", 20, 50)
    product3 = Product("Keyboard", 50, 30)

    
    inventory = Inventory()

    
    inventory.add_product(product1)
    inventory.add_product(product2)
    inventory.add_product(product3)

    
    inventory.display_available_products()

    
    inventory.update_product_quantity("Mouse", 45)

    
    inventory.display_available_products()
```

```python
class Shape:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)


rectangle = Shape(5, 4)

print(f"Area: {rectangle.calculate_area()}")
print(f"Perimeter: {rectangle.calculate_perimeter()}")
```

```python
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average_grade(self):
        return sum(self.grades) / len(self.grades)

    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {self.grades}")
        print(f"Average Grade: {self.calculate_average_grade()}")


student = Student(1, 'John', [85, 90, 78])

student.display_student_details()
```

```python
class UserProfile:
    def __init__(self, username, bio=""):
        """
        Initialize a user profile with a username and optional bio.
        """
        self.username = username
        self.bio = bio
        self.followers = 0
        self.posts = []

    def add_post(self, post_content):
        """
        Add a new post to the user's profile.
        """
        self.posts.append(post_content)

    def display_posts(self):
        """
        Display all posts in the user's profile.
        """
        print(f"Posts by {self.username}:")
        for post in self.posts:
            print(f"- {post}")

    def search_posts_by_keyword(self, keyword):
        """
        Search for posts containing the specified keyword.
        """
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if matching_posts:
            print(f"Posts containing '{keyword}':")
            for post in matching_posts:
                print(f"- {post}")
        else:
            print(f"No posts found containing '{keyword}'.")


user1 = UserProfile(username="john_doe", bio="Tech enthusiast")
user1.add_post("Just learned about Python decorators. So cool!")
user1.add_post("Exploring machine learning algorithms.")
user1.add_post("Enjoying a cup of coffee ☕️")

user1.display_posts()
user1.search_posts_by_keyword("Python")
```

```python
class ToDoList:
    def __init__(self):
        """
        Initialize an empty ToDo list.
        """
        self.tasks = {}  

    def add_task(self, task_name, due_date):
        """
        Add a new task to the list.
        Args:
            task_name (str): Name of the task.
            due_date (str): Due date for the task (e.g., '2024-03-20').
        """
        self.tasks[task_name] = due_date

    def mark_task_completed(self, task_name):
        """
        Mark a task as completed.
        Args:
            task_name (str): Name of the task to mark as completed.
        """
        if task_name in self.tasks:
            del self.tasks[task_name]
            print(f"Task '{task_name}' marked as completed.")
        else:
            print(f"Task '{task_name}' not found in the list.")

    def display_pending_tasks(self):
        """
        Display all pending tasks.
        """
        if self.tasks:
            print("Pending tasks:")
            for task, due_date in self.tasks.items():
                print(f"- {task} (Due on {due_date})")
        else:
            print("No pending tasks.")


my_todo_list = ToDoList()
my_todo_list.add_task("Buy groceries", "2024-03-18")
my_todo_list.add_task("Finish report", "2024-03-20")
my_todo_list.add_task("Call mom", "2024-03-19")

my_todo_list.display_pending_tasks()
my_todo_list.mark_task_completed("Buy groceries")
my_todo_list.display_pending_tasks()
```
