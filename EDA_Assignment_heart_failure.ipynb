```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
#¯´ What is the distribution of age among heart failure patients in the dataset
sns.displot(df['age'],kde = True)
```

```python
# the maximum age is distributed among 50-70
```

```python
# ´ How does the death rate vary with age
sns.swarmplot(y = 'age', x = 'DEATH_EVENT',data = df,hue = 'DEATH_EVENT',palette = 'coolwarm')
```

```python
# most of the death rates are among 50-75
```

```python
# What is the percentage of male and female patients in the dataset?
df['sex'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.figure(figsize = (8,8))
```

```python
# male is 64.9%
# female is 35.1%
```

```python
# How does the platelet count vary among different age groups<
sns.scatterplot(data = df, x = 'age', y = 'platelets')
```

```python
# max is from 45-70
```

```python
# Is there a correlation between creatinine and sodium levels in the blood<
selected_columns = ['serum_creatinine', 'serum_sodium']
df[selected_columns].corr()
sns.heatmap(df[selected_columns].corr(), annot = True )
```

```python
# there is a negative correlation between createnine and sodium levels
```

```python
#  How does the prevalence of high blood pressure differ between male and female patients<
sns.barplot(x = 'sex', y = 'high_blood_pressure',data = df)
```

```python
# females have more high pressure as compared to males
```

```python
#  What is the relationship between smoking habits and the occurrence of heart failure<
sns.jointplot(y = 'DEATH_EVENT', x = 'smoking',data = df,hue = 'DEATH_EVENT')
```

```python
# people who smoke have a higher chance of heart failure
```

```python
#  Are there any noticeable patterns in the distribution of death events across different age groups<

sns.scatterplot(x = 'DEATH_EVENT',y= 'age',data = df,hue = 'DEATH_EVENT')
```

```python
# lot of people are in the age group of 45-60
```

```python
# Is there any significant difference in ejection fraction between patients with and without diabetes<
sns.swarmplot(y='ejection_fraction',x = 'diabetes',data = df,hue = 'diabetes')
```

```python
# people not having diabetes have more ejection fraction
```

```python
# How does the serum creatinine level vary between patients who survived and those who did not?

sns.scatterplot(x = 'DEATH_EVENT', y = 'serum_creatinine',data = df,hue = 'DEATH_EVENT')
```

```python
# people who did not survived have more createnine levels
```
